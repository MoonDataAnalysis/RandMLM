#### 什么是网络方法（network approach）？网路方法可以回答什么问题？

网络方法用网络结构描述众多变量之间的关系，如下图：




以精神疾病举例，一种精神基本往往包含许多症状，这些症状之间有怎样的关系？是否某一种症状出现某种其他症状也会出现？

对于这种问题的回答，往往需要我们摒弃单路径的探讨，同时讨论多变量之间复杂路径关系。

#### 本课程介绍的可以对追踪数据描绘网络模型——Multilevel vector auto regression model

该方法既能探讨个体间的网络关系（例如，个体情绪水平越高，神经质也倾向越高），也能探讨个体内的网络关系（例如个体某晚谁的不好，情绪也会不好）

参考文献：Bringmann, L. F., Vissers, N., Wichers, M., Geschwind, N., Kuppens, P., Peeters, F., ... & Tuerlinckx, F. (2013). A network approach to psychopathology: new insights into clinical longitudinal data. PloS one, 8(4), e60188.
